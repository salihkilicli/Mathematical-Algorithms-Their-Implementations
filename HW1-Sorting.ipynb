## Quicksort

The quicksort algorithm is a divide and conquer algorithm originally developed in 1959 by Tony Hoare while in the Soviet Union, as a visiting student at Moscow State University.  This method effectively partitions the input data with respect to a pivot, and then recursively applies the algorithm to each subproblem defined by the pivot.

[Wikipedia](https://en.wikipedia.org/wiki/Quicksort) has an article describing the history and several variants of the algorithm.


## Algorithm Overview

- Divide: Choose a pivot, an element from the list, and partition the list so that all elements less than the pivot are in one partition and those greater appear in another partition.  
- Conquer: Recursively apply this to each sublist (partition)
- Combine: Combine the results. Typically this is trivial as the above steps are performed in place.





## Complexity

We commonly use asymptotic notation to describe the running time or space requirements of algorithms.  This notation is intended to describe the order of growth of an algorithm's running time (or space requriements), which in turn allows us to better understand the efficiency and relative performance of algorithms.


There are several types of asymptotic notation, the three most common are defined below (Big-Oh, $O(\cdot)$, is the most widely used):

$\Theta(g(n)) = \{f(n) : \exists\; c_1, c_2, n_0 > 0 \mid 0 \le c_1 g(n) \le f(n) \le c_2 g(n) \;\forall n \ge n_0 \}$ 

$O(g(n)) = \{f(n) : \exists\; c, n_0 > 0 \mid 0 \le f(n) \le c g(n) \;\forall n \ge n_0 \}$ 

$\Omega(g(n)) = \{f(n) : \exists\; c, n_0 > 0 \mid 0 \le c g(n) \le f(n)  \;\forall n \ge n_0 \}$ 



**Source**: *Introduction to Algorithms (3rd edition)* by Thomas H. Cormen, Charles E. Leiserson, Ronald L. Rivest, Clifford Stein, page 45, figure 3.1.

### Quicksort Time Complexity

There are many schemes for the selection of a pivot and partition.  The choice of which can greatly affect the overall algorithms performance. Below are the theoretical time complexities for the general algorithm. 

- Worst Case $O(n^2)$
- Best Case $\Omega(n\;log\;n)$
- Average Case $\Theta(n\;log\;n)$

## Lomuto Partition Scheme
<br>

<div STYLE="background-color:#000000; height:2px; width:100%;"></div>
<b>Algorithm:</b> Lomuto Partition
<div STYLE="background-color:#000000; height:1px; width:100%;"></div>

- <b>Input:</b> List $A$ of $n$ real numbers, left index $lo$, right index $hi$.
- <b>Output:</b> Rearrangement of the list $A$ and an index $i$ such that $A[j]\leq A[i] \iff j \leq i$.

<div STYLE="background-color:#000000; height:1px; width:100%;"></div>

1. $p \leftarrow A[hi]$
2. $i \leftarrow lo$
3. __for__ $j=lo$ to $hi-1$ __do__
4. $\quad$ __if__ $A[j]\leq p$ __then__
5. $\qquad$ Swap $A[i]$ with $A[j]$
6. $\qquad$ i = i + 1
7. $\quad$ __end if__
8. __end for__
9. Swap $A[i]$ with $A[hi]$
10. __return__ $i$

<div STYLE="background-color:#000000; height:2px; width:100%;"></div>

After running the above scheme once all elements in the (sub)list that are less than or equal to the pivot appear before the pivot, and all larger elements appear after the pivot.  Notice that the pivot is now in its final location in the original input list.  Also, by returning the index of the pivot (after it is put in its final location) we can operate on the sublists to the left and right of the pivot.

```python
def partition_lomuto(A, lo, hi):
    """Creates a lomuto partition for a given list
    A := A list of real numbers
    p := Pivot element = A[hi]
    """
    assert type(A) == list, "Type Error: A should be a list object"
    p = A[hi]                                           # Set pivot as the right element
    i = lo                                              # Set i to the left index

    for j in range(lo, hi):                             # Iterate list A left to right index (exclusive)
        if A[j] <= p:                                   # If A[j] <= p put on the left side of merged list
            A[i], A[j] = A[j], A[i]
            i += 1
        
    A[i], A[hi] = A[hi], A[i]                         
    return i
```

```python
A = [9, -3, 5, 2, 6, 8, -6, 1, 3]

print(f"pivot_index = {partition_lomuto(A, 0, len(A)-1)}")
print(f"l_partition = {A}")
```

## Hoare Partition Scheme

This scheme works by starting at each end of the (sub)list and working towards the center until an 'inversion' is found.  An inversion occurs when a smaller value than the pivot is found at the right index and a larger value than the pivot is found at the left index.  These values are swapped and the search continues. 

<br>

<div STYLE="background-color:#000000; height:2px; width:100%;"></div>
<b>Algorithm:</b> Hoare Partition
<div STYLE="background-color:#000000; height:1px; width:100%;"></div>

- <b>Input:</b> List A of $n$ real numbers.
- <b>Output:</b> Rearrangement of the list $A$ and an index $i$ such that $A[j]\leq A[i] \iff j \leq i$.

<div STYLE="background-color:#000000; height:1px; width:100%;"></div>

1. $p \leftarrow A[lo]$
2. $(i,\ j) \leftarrow (lo-1,\ hi+1)$
3. __while__ __do__
4. $\quad$ __while__ __do__
5. $\qquad j = j - 1$
6. $\qquad$ __if__ $A[j] \leq p$ __then__
7. $\quad \qquad$ __break__
8. $\qquad$ __end if__
9. $\quad$ __end while__
10. $\quad$ __while__ __do__
11. $\qquad i = i + 1$
12. $\qquad$ __if__ $A[i] \geq p$ __then__
13. $\quad \qquad$ __break__
14. $\qquad$ __end if__
15. $\quad$ __end while__
16. $\quad$ __if__ $i \geq j$ __then__
17. $\qquad$ __return__ j
18. $\quad$ __end if__
19. $\quad$ Swap $A[i]$ with $A[j]$
20. __end while__

<div STYLE="background-color:#000000; height:2px; width:100%;"></div>

Unlike the Lomuto scheme the pivot is not always placed in its final position after a single pass through the above scheme.  Instead the scheme guarantees that for a (sub)list with start index `a` and end index `b` the values in the (sub)list `[a,right]` will be less than the pivot, and those values in `[right+1, b]` will be greater than or equal to the pivot, where `right` is the index returned by the scheme.  Thus your implementation of quicksort should **not** assume that the returned index from the above scheme contains a value in its final location.

```python
def partition_hoare(A, lo, hi):
    """Creates a hoare partition for a given list
    A := A list of real numbers
    p := Pivot element = A[lo]
    """
    assert type(A) == list, "Type Error: A should be a list object"
    p = A[lo]
    i, j = lo-1, hi+1
    
    while True:
        while True:
            j -= 1
            if A[j] <= p:
                break              
        while True:
            i += 1
            if A[i] >= p:
                break              
        if i >= j:
            return j
        
        A[i], A[j] = A[j], A[i]
```

```python
A = [9, -3, 5, 2, 6, 8, -6, 1, 3]

print(f"pivot_index = {partition_hoare(A, 0, len(A)-1)}")
print(f"h_partition = {A}")
```

```python
def quicksort_lomuto(elements, low_idx=None, high_idx=None):
    """
    This function sorts the elements list in place using the 
    Lomuto partitioning scheme. Only the values in the range
    [low_idx, high_idx] are sorted by a call to this function.
    
    elements - A list of values to be sorted
    low_idx  - The starting index of the elements list to be 
               sorted.  Values appearing before this index are
               not touched. Default is 0.
    high_idx - The ending value of the elements list to be 
               sorted.  Values appearing after this index are
               not touched.  Default is len(elements)-1.
    """
    assert type(elements) == list, "Type Error: A should be a list object"
    if (low_idx == None) & (high_idx == None):
        low_idx, high_idx = 0, len(elements)-1
    
    if low_idx < high_idx:
        p = partition_lomuto(elements, low_idx, high_idx)
        quicksort_lomuto(elements, low_idx, p-1)
        quicksort_lomuto(elements, p+1, high_idx)
```

```python
import random, timeit, copy
import matplotlib.pyplot as plt  
import seaborn as sns
sns.set(style='darkgrid')
import time, sys
from IPython.display import clear_output


def update_progress(progress):
    """
    Helper function to display the progress of the tests.
    Source: https://www.mikulskibartosz.name/how-to-display-a-progress-bar-in-jupyter-notebook/
    """
    bar_length = 20
    if isinstance(progress, int):
        progress = float(progress)
    if not isinstance(progress, float):
        progress = 0
    if progress < 0:
        progress = 0
    if progress >= 1:
        progress = 1

    block = int(round(bar_length * progress))
    clear_output(wait = True)
    text = "Progress: [{0}] {1:.1f}%".format( "#" * block + "-" * (bar_length - block), progress * 100)
    print(text)

    
def plot_time_comparison(sample_func, num_repeats=None, 
                         max_size=None, step_size=None, 
                         lomuto=quicksort_lomuto,
                         hoare=quicksort_hoare):
    """
    Records and plots the runtimes of the quicksort_lomuto and 
    quicksort_hoare functions using input lists generated with 
    sample_func.  
    
    sample_func: Expected to take a single argument n and return
                 a length n list of values.
    num_repeats: The number of timing tests that will be average 
                 together for a single input list.
    max_size:    The maximum size list that will be considered 
                 by this function.  Sizes will be tested in the
                 range [1, max_size]
    step_size:   The increment value to be used while testing.
                 If step_size=1 then all list lengths in the 
                 range [1, max_size] will be tested.  Instead
                 if step_size=10, then the lengths 1, 11, 21,
                 ... will be tested until max_size is reached
                 or exceeded.
    """
    # The number of times the quicksort function will be called on a
    # given input list.  The timing results will be averaged over these
    # repeated executions.
    if num_repeats is None:
        num_repeats = 10

    # The maximum size list to consider
    if max_size is None:
        max_size = 1000

    # The step to be used - compute time for every `step_size` values
    if step_size is None:
        step_size = 10

    # Storage for the time data for each test below
    time_lomuto = []
    time_hoare = []

    # Time the sorting of random lists with sizes ranging from 1 
    # to max_size
    for n in range(1, max_size, step_size):

        # Create our random list of length n
        elements = sample_func(n)
        # Create a copy of the input list to ensure that the second
        # method gets the same input data and not a pre-sorted list
        elements2 = copy.deepcopy(elements)

        # Lomuto Timing Test
        timer = timeit.Timer(lambda: lomuto(elements, 0, n-1))
        t = timer.timeit(number = num_repeats)
        time_lomuto.append(t / num_repeats)

        # Hoare Timing Test
        timer = timeit.Timer(lambda: hoare(elements2, 0, n-1))
        t = timer.timeit(number = num_repeats)
        time_hoare.append(t / num_repeats)

        # Update the progress bar
        update_progress((n-1) / max_size)
    update_progress(1)

    plt.figure(figsize=(8, 6))
    plt.plot(range(1, max_size, step_size), time_lomuto, 'b--')
    plt.plot(range(1, max_size, step_size), time_hoare, 'g--')
    plt.title('Time complexity comparison of Lomuto and Hoare quicksort methods')
    plt.xlabel('Step size')
    plt.ylabel('Time')
    plt.tick_params(axis='y', which='both', labelleft='on', labelright='on')
    plt.legend(['Lomuto', 'Hoare'], loc='best')
    plt.show()
```

## Your Solution (3)

Is there a difference between the runtimes of these two schemes? If so, why? 

__Answer:__ As it can be seen from the graph above, Hoare's quicksort algorithm converges much faster than Lomuto method. This is mainly because the number of swaps on average to make a partition (for a list of $n$ elements) in two different algorithms:

- Lomuto's algorithm swaps on average: $\quad \dfrac{n}{2} - \dfrac{1}{2}$
 
- Hoare's algorithm swaps on average: $\quad \dfrac{n}{6} - \dfrac{1}{3}$
 
As a result, Hoare's partition works much faster and hence the quicksort algorithm itself. Moreover, in the case of equal number of elements for both algorithms, __Hoare__ $\approx \mathcal{O}(nlog(n))$ whereas __Lomuto__ $\approx \Theta(n^2)$.
